Import the libraries

```python
%matplotlib inline  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import seaborn as sns
```

Import the dataset from the file system

```python
#India_dataset = 'Dataset/India_Data/API_IND_DS2_en_csv_v2_10135894/API_IND_DS2_en_csv_v2_10135894.csv'
GDP_dataset = 'Dataset/GDP/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10134290/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10134290.csv'
HDI_dataset = 'Dataset/HDI/Human Development Index (HDI).csv'
Infant_Mortality_dataset = 'Dataset/Infant_Mortality_Rate/API_SP.DYN.IMRT.IN_DS2_en_csv_v2_10181276/API_SP.DYN.IMRT.IN_DS2_en_csv_v2_10181276.csv'
Life_Expectancy_dataset = 'Dataset/Life_Expectancy/API_SP.DYN.LE00.IN_DS2_en_csv_v2_10181296/API_SP.DYN.LE00.IN_DS2_en_csv_v2_10181296.csv'
```

Define a function which returns a list of 33 developed countries as per per CIA (https://www.cia.gov/library/publications/the-world-factbook/appendix/appendix-b.html#D)

NOTE: 
1. Taiwan data doesnt exist in the world bank database. So, ommitted.
2. Hong Kong data doesnt exist in the world bank database for Infant Mortality Rate. So, ommitted.


```python
def developed_countries_list():
    dc_list = []
    dc_list.append("Australia")
    dc_list.append("Austria")
    dc_list.append("Belgium")
    dc_list.append("Canada")
    dc_list.append("Cyprus")
    dc_list.append("Czech Republic")
    dc_list.append("Denmark")
    dc_list.append("Finland")
    dc_list.append("France")
    dc_list.append("Germany")
    dc_list.append("Greece")
    #dc_list.append("Hong Kong SAR, China")
    dc_list.append("Iceland")
    dc_list.append("Ireland")
    dc_list.append("Israel")
    dc_list.append("Italy")
    dc_list.append("Japan")
    dc_list.append("Korea, Rep.")
    dc_list.append("Luxembourg")
    dc_list.append("Malta")
    dc_list.append("Netherlands")
    dc_list.append("New Zealand")
    dc_list.append("Norway")
    dc_list.append("Portugal")
    dc_list.append("Singapore")
    dc_list.append("Slovak Republic")
    dc_list.append("Slovenia")
    dc_list.append("Spain")
    dc_list.append("Sweden")
    dc_list.append("Switzerland")
    #dc_list.append("Taiwan")
    dc_list.append("United Kingdom")
    dc_list.append("United States")
    return dc_list
```

Get the dataframes from the csv files

```python
gdp_df = pd.read_csv(GDP_dataset,skiprows=3)
hdi_df = pd.read_csv(HDI_dataset,skiprows=1)
im_df = pd.read_csv(Infant_Mortality_dataset,skiprows=3)
le_df = pd.read_csv(Life_Expectancy_dataset,skiprows=3)
```

For the **hdi_df** dataframe:
1. Filter NAN columns which are alternate
2. Remove the space from the name of the countries in the Country column
3. Rename the countries which have different name in the other dataframes to make data same throughout
4. Add INDIA to the list of countries so as to make an analysis with other countries.

```python
# Filter NAN columns which are alternate
hdi_drop_col = []
for i in range(3,hdi_df.shape[1]):
    if(i%2 == 1):
        hdi_drop_col.append(i)
hdi_df = hdi_df.drop(hdi_df.columns[hdi_drop_col], axis=1)

# Remove initial spaces from the countries names 
# AND
# Rename country names list and make it similar to the World Bank dataset
for i in range(0,hdi_df['Country'].shape[0]):
    s = hdi_df['Country'][i]
    if(s[1:] == "Czechia"):
        hdi_df['Country'][i] = "Czech Republic"
    #elif(s == "Hong Kong, China (SAR)"):
    #    s = "Hong Kong SAR, China"
    elif(s[1:] == "Korea (Republic of)"):
        hdi_df['Country'][i] = "Korea, Rep."
    elif(s[1:] == "Slovakia"):
        hdi_df['Country'][i] = "Slovak Republic"
    else:
        hdi_df['Country'][i] = s[1:]

# Get the list of all developed countries   
countries_list = developed_countries_list()
# Add India among other countries for comparison and analysis
countries_list.append("India")
```

Filter all the dataframes with the countries that are in the list : countries_list

```python
developed_gdp_df = gdp_df[gdp_df['Country Name'].isin(countries_list)]
developed_im_df = im_df[im_df['Country Name'].isin(countries_list)]
developed_le_df = le_df[le_df['Country Name'].isin(countries_list)]
developed_hdi_df = hdi_df[hdi_df['Country'].isin(countries_list)]
```

Filter GDP data from 1995 - 2016 <br/>
Normalize the GDP data row wise

```python
#Filter
developed_gdp_df_filtered = developed_gdp_df.drop(developed_gdp_df.columns[1:39], axis=1)
developed_gdp_final_df = developed_gdp_df_filtered.drop(developed_gdp_df_filtered.columns[[developed_gdp_df_filtered.shape[1]-2,developed_gdp_df_filtered.shape[1]-1]], axis=1)
developed_gdp_final_df = developed_gdp_final_df.reset_index(drop=True)

#Normalize
developed_gdp_final_rowmax = developed_gdp_final_df.select_dtypes(include='float64').max(axis=1)
for i in range(0,developed_gdp_final_df.shape[0]):
    rowmax = developed_gdp_final_rowmax[i]
    for j in range(0,developed_gdp_final_df.columns[1:].shape[0]):
        data = developed_gdp_final_df.loc[i, developed_gdp_final_df.columns[1:][j]]
        data = data/rowmax
        developed_gdp_final_df.loc[i, developed_gdp_final_df.columns[1:][j]] = data
```

Filter Infant Mortality Rate data from 1995 - 2016 <br/>
Normalize the Infant Mortality Rate data row wise

```python
#Filter
developed_im_df_filtered = developed_im_df.drop(developed_im_df.columns[1:39], axis=1)
developed_im_final_df = developed_im_df_filtered.drop(developed_im_df_filtered.columns[[developed_im_df_filtered.shape[1]-2,developed_im_df_filtered.shape[1]-1]], axis=1)
developed_im_final_df = developed_im_final_df.reset_index(drop=True)

#Normalize
developed_im_final_rowmax = developed_im_final_df.select_dtypes(include='float64').max(axis=1)
for i in range(0,developed_im_final_df.shape[0]):
    rowmax = developed_im_final_rowmax[i]
    for j in range(0,developed_im_final_df.columns[1:].shape[0]):
        data = developed_im_final_df.loc[i, developed_im_final_df.columns[1:][j]]
        data = data/rowmax
        developed_im_final_df.loc[i, developed_im_final_df.columns[1:][j]] = data
```

Filter Life Expectancy data from 1995 - 2016 <br/>
Normalize the Life Expectancy data row wise

```python
#Filter
developed_le_df_filtered = developed_le_df.drop(developed_le_df.columns[1:39], axis=1)
developed_le_final_df = developed_le_df_filtered.drop(developed_le_df_filtered.columns[[developed_le_df_filtered.shape[1]-2,developed_le_df_filtered.shape[1]-1]], axis=1)
developed_le_final_df = developed_le_final_df.reset_index(drop=True)

#Normalize
developed_le_final_rowmax = developed_le_final_df.select_dtypes(include='float64').max(axis=1)
for i in range(0,developed_le_final_df.shape[0]):
    rowmax = developed_le_final_rowmax[i]
    for j in range(0,developed_le_final_df.columns[1:].shape[0]):
        data = developed_le_final_df.loc[i, developed_le_final_df.columns[1:][j]]
        data = data/rowmax
        developed_le_final_df.loc[i, developed_le_final_df.columns[1:][j]] = data        
```

Filter Human Development Index data from 1995 - 2016 <br/>
Normalize the Human Development Index data row wise is already done in the dataset

```python
#Filter
developed_hdi_df_filtered = developed_hdi_df.drop(developed_hdi_df.columns[2:7], axis=1)
developed_hdi_df_filtered = developed_hdi_df.drop(developed_hdi_df.columns[[developed_hdi_df.shape[1]-1]], axis=1)
developed_hdi_final_df = developed_hdi_df_filtered.drop(developed_hdi_df_filtered.columns[0], axis=1)
developed_hdi_final_df = developed_hdi_final_df.reset_index(drop=True)
```

Sort all the Dataframes

```python
developed_gdp_sorted_df = developed_gdp_final_df.sort_values(by=['Country Name'])
developed_im_sorted_df = developed_im_final_df.sort_values(by=['Country Name'])
developed_le_sorted_df = developed_le_final_df.sort_values(by=['Country Name'])
developed_hdi_sorted_df = developed_hdi_final_df.sort_values(by=['Country'])
```

```python
final_df = pd.DataFrame().reindex_like(developed_gdp_sorted_df)
for i in range(0,developed_gdp_sorted_df.shape[0]):
    final_df.loc[i, final_df.columns[0]] = developed_gdp_sorted_df.loc[i,developed_gdp_sorted_df.columns[0]]
    for j in range(0,developed_gdp_sorted_df.columns[1:].shape[0]):
        data = (developed_gdp_sorted_df.loc[i, developed_gdp_sorted_df.columns[1:][j]] + developed_im_sorted_df.loc[i, developed_im_sorted_df.columns[1:][j]] + developed_le_sorted_df.loc[i, developed_le_sorted_df.columns[1:][j]] + developed_hdi_sorted_df.loc[i, developed_hdi_sorted_df.columns[1:][j]])/4
        final_df.loc[i, final_df.columns[1:][j]] = data
```
